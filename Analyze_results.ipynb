#### Adjucancy matrices represented as directed graphs to show causality among time series

```python
# Example adjacency matrix as a Pandas DataFrame

# Specify font color for node names
font_color = 'black'

# Specify node size
node_size = 500

for adjacency_matrix in res_list:
    np.fill_diagonal(adjacency_matrix.values, 0)
    minVal = 0 #adjacency_matrix.unstack().quantile(0.75)
    #adj = 1*(adjacency_matrix>minVal)
    adj = adjacency_matrix.T
    np.fill_diagonal(adj.values, 0)
    # Create a directed graph from the adjacency matrix
    G = nx.from_pandas_adjacency(adj, create_using=nx.DiGraph())
    node_colors = 'palegreen'
    # Plot the directed graph
   # pos = nx.star_layout(G)  # You can choose a different layout if needed
    #pos = nx.circular_layout(G)
    pos = nx.shell_layout(G)
    #pos = nx.kamada_kawai_layout(G)
    #pos = nx.spectral_layout(G)
    offset = 0.5
    pos_adjusted = {node: (x + offset, y) for node, (x, y) in pos.items()}
    nx.draw(G, pos=pos_adjusted, with_labels=True, node_color = node_colors, node_size = node_size , arrowsize=20,
           edge_color='darkgreen',  # Change arrow color
        width=0.5,           # Change arrow thickness
        connectionstyle='arc3,rad=0.2',  # Change arrow shape
        arrowstyle= '->' ) #    # Change arrow style) #font_weight='bold'
    plt.axis('off')
    axis = plt.gca()
    axis.set_xlim([1.2*x for x in axis.get_xlim()])
    axis.set_ylim([1.2*y for y in axis.get_ylim()])
    #plt.tight_layout()
    plt.show()
```
