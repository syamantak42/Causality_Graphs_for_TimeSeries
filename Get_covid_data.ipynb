```python
import pandas as pd
import requests
import numpy as np
from io import StringIO

```

```python
def download_covid_data(url):
    response = requests.get(url)
    content = response.content.decode('utf-8')
    df = pd.read_csv(StringIO(content))
    return df

# URL for the JHU CSSE city-level dataset
jhu_covid_url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv"

# Download the dataset
covid_data = download_covid_data(jhu_covid_url)

# Print the first few rows of the dataset
display(covid_data.head())
```

##### Get covid data for US 

```python
df = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_US.csv")

# Specify the columns to be dropped
columns_to_drop = ['UID', 'iso2', 'iso3', 'code3', 'FIPS', 'Admin2',  'Country_Region', 'Lat', 'Long_', 'Combined_Key']

# Drop the specified columns
df = df.drop(columns=columns_to_drop)
df1 = df.groupby("Province_State").agg("sum").reset_index().transpose()
df1.reset_index().head()


df1 = df1.rename_axis("date").reset_index()
df1 = df1.rename_axis(None, axis=1)
df1['date'] = pd.to_datetime(df1['date'], format='%m/%d/%y')

# Filter entries after March 2020
filtered_df = df1[df1['date'] >= '2020-04-01']

filtered_df.index = filtered_df["date"]
filtered_df =filtered_df.drop(columns = ["date"])
filtered_df.head()
filtered_df.columns
filtered_df =filtered_df.drop(columns = ['American Samoa', 'Virgin Islands', 'Diamond Princess', 'Grand Princess','Guam',   'Northern Mariana Islands'])
```
